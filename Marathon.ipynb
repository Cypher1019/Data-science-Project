```python
#Year of event                  int64
#Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed        float64
#Athlete ID                     int64
#athlete_age                    int64
```

```python
df2 = df2.rename(columns = { 'Year of event':'year',
                             'Event dates':'race_day',
                             'Event name':'race_name',
                             'Event distance/length':'race_length',
                             'Event number of finishers':'race_number_of finishers',
                             'Athlete performance':'athlete_performance',
                             'Athlete gender':'athlete_gender',
                             'Athlete average speed': 'athlete_avergae_speed',
                             'Athlete ID': 'athlete_id'
                           })
```

```python
df3 = df2[['race_day', 'race_name', 'race_length', 'race_number_of finishers', 'athlete_id', 'athelete_gender', 'athlete_age', 'athlete_performance', 'athlete_avergae_speed', 'year' ]]
```

```python
#what age groups are the best in the 50mi race(20 + races min)(show 15)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_avergae_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count > 19').head(15)
```

```python
#what age groups are the worst in the 50mi race(20 + races min)(show 20)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_avergae_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count > 19').head(20)
```
